```python
# control flow
# aritmetika
a = 10
b = 5
print(a + b) # operator tambah (+)
print(a - b) # operator kurang (-)
print(a / b) # operator bagi (/)
print(a * b) # operator kali (*)
print(a ** b) # operator pangkat (**)
print(a % b) # operator sisa hasil bagi/modulus(%)
```

```python
phi = 3.14
r = 7
luas_lingkaran = phi * (r ** 2)
print("Luas lingkaran: ", luas_lingkaran)
```

```python
# Operator comparison
print(a > b) #True
print(a < b) #False
print(a >= b) # True
print(a <= b) # False
print(a == b) # False
print(a != b) # True
```

```python
# Logical operator (and , or , not)
print(a > b and a >= b) # true dan true
print(a > b and a < b) # true dan false
print(a > b or a < b) # true atau false
print(not(a > b and a >= b)) #ini akan false
```

```python
# Control Flow If

"""
if (kondisi => True):
  blok kode yang akan dieksekusi
"""

nilai_siswa = 90

if nilai_siswa >= 90: # ini True
  print("Grade A")
```

```python
# if else
nilai_siswa = 85

if nilai_siswa >= 90:
  print("Grade A")
else:
  print("Kurang Gacor")
```

```python
# if elif else
nilai_siswa = int(input("Masukkan Nilai Kamu: "))

if nilai_siswa >= 90:
  print("Grade A")
elif nilai_siswa < 90 and nilai_siswa >= 80:
  print("Grade B")
elif nilai_siswa < 80 and nilai_siswa >= 70:
  print("Grade C")
else:
  print("kurang gacorr")
```
